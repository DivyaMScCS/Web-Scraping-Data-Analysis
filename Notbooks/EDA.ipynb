# Products Analysis

#### a) What is the average price of products, and how does it vary across categories and materials?

```python

avg_price = df['Price'].mean()
print(f"Overall Average Product Price: {avg_price:.2f}")
```

```python
avg_price_category = df.groupby("Category")["Price"].mean().reset_index()
print("\nAverage Price by Category:")
display(avg_price_category)

avg_price_material = df.groupby("Material")["Price"].mean().reset_index()
print("\nAverage Price by Material:")
display(avg_price_material)
```

#### b. Is there a correlation between MRP and discount percentage, and do higher-MRP products receive steeper discounts?

```python

df["Discount_Percentage"] = ((df["MRP_Rate"] - df["Price"]) / df["MRP_Rate"]) * 100
correlation_mrp_discount = df[["MRP_Rate", "Discount_Percentage"]].corr().iloc[0, 1]
print(f"Correlation between MRP and Discount Percentage: {correlation_mrp_discount:.2f}")
if correlation_mrp_discount > 0:
    print("   â†’ Higher-MRP products tend to receive higher discounts.")
else:
    print("   â†’ No strong relationship between MRP and discount percentage.")
```

```python
plot3 = ggplot(df, aes(x='MRP_Rate', y='Discount_Percentage')) + geom_point(color='red') + ggtitle("MRP vs Discount Percentage")
plot3.show()
```

#### c. How many products have zero ratings or reviews, and what factors contribute to low customer engagement?

```python

zero_ratings = df[df["Rating"] == 0].shape[0]
zero_reviews = df[df["Reviews"] == 0].shape[0]
print(f"Products with Zero Ratings: {zero_ratings}")
print(f"Products with Zero Reviews: {zero_reviews}")

plot = ggplot(df, aes(x='Rating')) + \
    geom_bar(fill='blue') + \
    ggtitle("Distribution of Product Ratings")
plot.show()

```

```python
df['Reviews'] = df['Reviews'].astype(str) 

plot2 = ggplot(df, aes(x='Reviews')) + geom_bar(fill='pink') + ggtitle("Distribution of Product Reviews")
plot2.show()
```

#### d. What is the relationship between product price and the number of reviews, and do expensive products attract more customer feedback?

```python
correlation_price_reviews = df[["Price", "Reviews"]].corr().iloc[0, 1]
print(f"Correlation between Price and Number of Reviews: {correlation_price_reviews:.2f}")
if correlation_price_reviews > 0:
    print("   â†’ Higher-priced products tend to receive more reviews.")
else:
    print("   â†’ No strong relationship between price and reviews.")

```

```python
plot5 = ggplot(df, aes(x='Price', y='Reviews')) + geom_point(color='blue') + ggtitle("Price vs Number of Reviews")
plot5.show()
```

#### e. Are specific product categories more likely to have longer or shorter estimated delivery times, and what factors influence this?

```python
df["Estimated_Delivery"] = pd.to_datetime(df["Estimated_Delivery"])
avg_delivery_time_category = df.groupby("Category")["Estimated_Delivery"].mean()
print("\n5. Average Estimated Delivery Time by Category:")
print(avg_delivery_time_category)
```

#### f. Which materials are most commonly used in products?

```python
material_counts = df["Material"].value_counts().reset_index()
material_counts.columns = ["Material", "Count"]

# Display results
print("Most Commonly Used Materials in Products:")
print(material_counts)
```

# Price Analysis

#### a. What are the highest and lowest priced products, and do they align with customer ratings?

```python

highest_price_product = df.loc[df["Price"].idxmax(), ["Product_Name", "Price", "Rating", "Reviews"]]
lowest_price_product = df.loc[df["Price"].idxmin(), ["Product_Name", "Price", "Rating", "Reviews"]]


result_df = pd.DataFrame([highest_price_product, lowest_price_product], index=["Highest Priced Product", "Lowest Priced Product"])


styled_df = result_df.style.apply(lambda x: ["background-color: yellow" if x.name == "Highest Priced Product" else "background-color: lightgreen" for i in x], axis=1)
styled_df.set_properties(**{'text-align': 'left'})
styled_df.set_table_styles([{'selector': 'th', 'props': [('font-size', '14pt')]}, {'selector': 'td', 'props': [('font-size', '12pt')]}])
```

#### b. What is the price distribution of products with zero ratings or reviews?

```python
zero_rating_reviews = df[(df["Rating"] == 0) | (df["Reviews"] == 0)]["Price"].describe()
print("\n2. Price Distribution of Products with Zero Ratings or Reviews:")
zero_rating_reviews
```

#### c. Do higher-priced products have longer or shorter estimated delivery times?

```python

df['Estimated_Delivery'] = pd.to_datetime(df['Estimated_Delivery'])

reference_date = pd.to_datetime('2025-03-01')
df['Delivery_Days'] = (df['Estimated_Delivery'] - reference_date).dt.days

correlation_price_delivery = df[["Price", "Delivery_Days"]].corr().iloc[0, 1]
print(f"\n Correlation between Price and Estimated Delivery Time: {correlation_price_delivery:.2f}")
if correlation_price_delivery > 0:
    print("   â†’ Higher-priced products tend to have longer delivery times.")
else:
    print("   â†’ Higher-priced products are delivered faster.")
```

#### d. Do higher-priced products have better ratings and more customer reviews?

```python
correlation_price_rating = df[["Price", "Rating"]].corr().iloc[0, 1]
correlation_price_reviews = df[["Price", "Reviews"]].corr().iloc[0, 1]
print(f"\n  Correlation between Price and Ratings: {correlation_price_rating:.2f}")
print(f"   Correlation between Price and Number of Reviews: {correlation_price_reviews:.2f}")
if correlation_price_rating > 0:
    print("   â†’ Higher-priced products tend to have better ratings.")
else:
    print("   â†’ No strong relationship between price and ratings.")
```

#### e. How do price variations impact customer engagement (ratings & reviews)?

```python

df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')

df["Price_Range"] = pd.qcut(df["Price"], q=4, labels=["Low", "Mid-Low", "Mid-High", "High"])

price_engagement = df.groupby("Price_Range")[["Rating", "Reviews"]].mean()

styled_df = price_engagement.style.apply(lambda x: ["background-color: yellow" if x.name == "Low" else
                                                    "background-color: lightgreen" if x.name == "Mid-Low" else
                                                    "background-color: lightblue" if x.name == "Mid-High" else
                                                    "background-color: lightcoral" for i in x], axis=1)


styled_df.set_properties(**{'text-align': 'left'})
styled_df.set_table_styles([{'selector': 'th', 'props': [('font-size', '14pt')]}, {'selector': 'td', 'props': [('font-size', '12pt')]}])
```
